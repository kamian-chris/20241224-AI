````python
# Step0. import...
import urllib.request as req
import bs4 as bs
import pandas as pd

# Step1. 先準備空的 list[]
table = []

page = 1
while True:
    print("頁數:", page)
    url = "https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt".format(page)
    page = page + 1
    response = req.urlopen(url)
    html = bs.BeautifulSoup(response)

    r_list = html.find_all("div", {"class":"list-rst__body"})
    if len(r_list) == 0:
        break

    for r in r_list:
        en_name = r.find("a", {"class":"list-rst__rst-name-target"})
        en_name_href = en_name["href"]
        en_name_text = en_name.text
        area = r.find("div", {"class":"list-rst__area-genre"})
        area_text = area.text.strip()
        intro = r.find("p", {"class":"list-rst__pr-title"})
        if intro == None:
            intro_text = "無"
        else:
            intro_text = intro.text.strip()
        rating = r.find("span", {"class":"c-rating__val"})
        rating_text = rating.text
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        dinner_price = prices[0].text
        lunch_price = prices[1].text
        holiday = r.find("span", {"class":"list-rst__holiday-text"})
        holiday_text = holiday.text

        print(rating_text)
        print(en_name_text)
        print(en_name_href)
        print(area_text)
        print(intro_text)
        print(holiday_text)
        print("晚間價錢:", dinner_price, "午間價錢:", lunch_price)
        print("-" * 30)

        # Step2. 每一筆資料就是一個字典( 準備字典{} )
        data = {
            "name(en)": en_name_text,
            "url": en_name_href,
            "area": area_text,
            "intro": intro_text,
            "rating": rating_text,
            "holiday": holiday_text
        }
        table.append(data)

# Step3. 轉乘DateFrame儲存

df = pd.DataFrame(table,
          columns=["name(en)", "url", "area", "intro", "rating", "holiday"])
df.to_csv("tabelog.csv", encoding="utf-8")
````

````python
# 讀取檔案
import pandas as pd
df = pd.read_csv("tabelog.csv", encoding="utf-8")
# 檢查 df["area"]

def func(s):
  real_area = s.split("/")[0]
  return real_area.strip()
  # 也可以 real_area = s.split(" / ")[0]
  #     return real_area

# 檢查 df["area"]
# 下面回傳
df["area"] = df["area"].apply(func)
# 數 area 數量
df["area"].value_counts()
````

````python
# 我先用自訂函式，來將 字串 轉換為 True/False (因為之後 條件 篩選需與 index 數量相等)
def func(s):
  if "惠比壽車站" in s:
    return True
  else:
    return False
# 然後用這個 新函式，跟別的條件，創造 fil(篩選)(篩選需與 index 數量相等)
fil = (df["area"].apply(func)) & (df["rating"].astype("float") >= 3.9)
# 最後用 df(整個文件)[ fil(篩選) ]
df[fil]
````

````python
# s 是值, name 是參數
def func(s, name):
  if name in s:
    return True
  else:
    return False

# 在.apply(func, 加值 -> args=["字串"](記得用中括號) )
fil = df["area"].apply(func, args=["二子玉川車站"])
df[fil]
````

# 抓出關鍵詞
```python
1. 西方型: 英文 字母->詞彙->片語(phrase)
2. 東方型: 中文     字->詞彙

分詞: 我喜歡吃甜食 -> 我/喜歡/吃/甜食

哪個詞彙在整段話裡最重要:
1. 出現次數(TF): 高, 重要性越高
2. 常用程度(IDF): 低, 重要性越高

詞彙重要程度:
TF-IDF = 出現次數 * (1/常用程度)

!!! jieba函式庫: 預先統計了常用程度
DF = 詞彙出現在幾篇裡面/夠多篇文章
!!!      : 分詞
利用詞彙庫做出初步劃分->(沒在詞彙庫/混淆)上下文推測哪種分法最合理

```

````python
# 以下都是設置
import jieba
import urllib.request as req
url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
req.urlretrieve(url, "dict")
jieba.set_dictionary('dict')

# 創造自己的 txt -> 名詞 + (強度 -> 越高越有機會被採用 -> 如 999999) +屬性(n, ns,....)
# 上傳再拿進來 ->使用 jieba.load_userdict("該 txt 檔名")
extra_url = "https://github.com/Elwing-Chou/tibame1224/raw/refs/heads/main/extra.dict.txt"
req.urlretrieve(extra_url, "extra.dict.txt")
jieba.load_userdict("extra.dict.txt")
````

````python
# 用空白建(" ")做分割，再join()進來
" ".join(jieba.cut(news))
````

````python
# 這邊才開始找跟顯示
import jieba.analyse
# allowPOS=["屬性"] -> 找對應屬性的詞
jieba.analyse.extract_tags(news, allowPOS=["n", "nr", "ns", "nt", "nz"])
````

````python
# 使用 jieba 來看一下這些評論到底哪些詞彙最重要
intro_all = "\n".join(df["intro"])
jieba.analyse.extract_tags(intro_all, allowPOS=["n", "nr", "ns", "nt", "nz"])
````
